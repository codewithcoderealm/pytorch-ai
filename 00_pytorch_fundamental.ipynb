## Reshaping, stacking, sequeezing and unsequeezing tensors

* Reshaping- reshapes an input tensor to defined shape
* stacking - combine multiple tensors
* sequeeze - remove all
* unsequeeze - add a `1` dimension to a target tensor
* permute

```python
import torch
```

```python
x= torch.arange(1.,10.)
x, x.shape
```

```python
# Add an extra dimnesion
x_reshaped = x.reshape(9,1)
x_reshaped,x_reshaped.shape
```

```python
x = torch.arange(1., 10.)
x, x.shape
```

```python
# Change the view
z = x.view(1,9)
z,z.shape
```

```python
# Changing z changes x
z[:,0]=5
z,x
```

```python
# Stack tensors on top of each other
x_stacked = torch.stack([x,x,x,x],dim=0)
x_stacked
```

```python
# torch.seqeeze()
print(f"Previous tensor: {x_reshaped}")
print(f"Previous shape: {x_reshaped.shape}")

# Remove extra dimensions from x_shaed
x_squeezed = x_reshaped.squeeze()
print(f"\nNew tensor: {x_squeezed}")
print(f"New shape: {x_squeezed.shape}")
```

```python
# torch.unsqueeze() - adds s single dimension to target tensor at specific dim
print(f"Previous target: {x_squeezed}")
print(f"Previous shape: {x_squeezed.shape}")

# Add an extra dimension with unsqueeze
x_unsqueezed = x_squeezed.unsqueeze(dim=0)
print(f"\nNew tensor: {x_unsqueezed}")
print(f"New shape: {x_unsqueezed.shape}")
```

```python
# torch.permute - reaarages the dimensions of a target tensor in sepcified order
x_original = torch.rand(size=(224,224,3))
# permute the oriignal tensor to rearrage the axis order
x_permuted = x_original.permute(2,0,1) #shifts axis0_>1,1->2,2->0
print(f"Previous shape: {x_original.shape}")
print(f"New shape: {x_permuted.shape}")
```

## PyTorch tensors & NumPy

NumPy is popular scientific Python numerical computing library

And beacause of this, PyTorch has funcationality to interact with it.

* Data in MumPy, want inPyTorch tensor -> `torch.from_numpy `(ndArray)`
* PyTorch tensor -> NumPy -> `torch.NumPy.Tensor`

```python
# NumPy array to tensor

import torch
import numpy as np

array = np.arange(1.0,8.0)
tensor = torch.from_numpy(array).type(torch.float32)
array,tensor
```

```python
# Change the value of array, what will this do to `tensor`
array = array+1
array,tensor
```

```python
# Tensor to NumPy array
tensor = torch.ones(7)
numpy_tensor = tensor.numpy()
tensor,numpy_tensor
```

```python
tensor = tensor + 1
tensor,numpy_tensor
```

## Reproducibility (trying to take random out of random)

In short how a neural network learns:
`start with random number -> tensor operations -> update random number to try and make them of the data -> again -> again -> again`

To reduce the randomness in neural network and PyTorch comes the concept of a **random seed**.

Essentially what the random seed does is "flavour" the radomness

```python
import torch
# Create two random tensors
random_tensor_A = torch.rand(3,4)
random_tensor_B = torch.rand(3,4)

print(random_tensor_A)
print(random_tensor_B)

print(random_tensor_A == random_tensor_B)
```

```python
# Lets make some random but reproducible tensors

import torch
# Set the radnom seed
RANDOM_SEED = 42

torch.manual_seed(RANDOM_SEED)
random_tensor_C = torch.rand(3,4)

torch.manual_seed(RANDOM_SEED)
random_tnsor_D = torch.rand(3,4)

print(random_tensor_C)
print(random_tnsor_D)
print(random_tensor_C == random_tnsor_D)

```
