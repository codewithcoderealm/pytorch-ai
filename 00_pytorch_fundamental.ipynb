### manipulating Tensors (tensors operations)

Tensor operations include:
* Addition
* Subtraction
* Multiplication (element-wise)
* Division
* Matrix Multiplication

```python
# Create a tensor
tensor = torch.tensor([1,2,3])
tensor + 10
```

```python
# Mulitply tensor by 10
tensor * 10
```

```python
# Subtract 10
tensor - 10
```

```python
# Try out PyTorch in-built functions
torch.mul(tensor,10)
```
